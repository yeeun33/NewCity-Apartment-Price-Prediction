# **CPI(소비자물가지수) 변수 생성**

### 1. 라이브러리 임포트

```python
import pandas as pd
import os
```

### 2. 연도별 CPI 데이터 정의

```python
# 한국은행 소비자물가지수 총지수 (2020년 = 100 기준)
# 출처: 통계청 소비자물가지수(CPI) 공식 누리집(https://mods.go.kr/cpi/)
CPI_DATA = {
    2006: 76.081,
    2007: 78.010,
    2008: 81.656,
    2009: 83.906,
    2010: 86.373,
    2011: 89.850,
    2012: 91.815,
    2013: 93.010,
    2014: 94.196,
    2015: 94.861,
    2016: 95.783,
    2017: 97.645,
    2018: 99.086,
    2019: 99.466,
    2020: 100.000,
    2021: 102.500,
    2022: 107.720,
    2023: 111.590,
    2024: 114.180,
    2025: 116.610,
    2026: 119.920,  # 5월 기준
}
```

### 3. CPI 파생변수 생성 함수 정의

```python
def add_CPI(df):
    # 계약년월 앞 4자리에서 계약연도 추출 (예: 202605 → 2026)
    df['계약연도'] = df['계약년월'].apply(lambda x: int(str(x)[:4]))

    # 계약연도를 기준으로 CPI 딕셔너리에서 값을 조회해 열 추가
    df['CPI'] = df['계약연도'].apply(lambda x: CPI_DATA.get(x))

    # CPI 값이 없는 행이 있는지 확인
    missing_mask = df['CPI'].isnull()
    if missing_mask.sum() > 0:
        # 누락된 연도 목록 출력
        missing = pd.unique(df.loc[missing_mask, '계약연도'])
        print(f"CPI 데이터 없는 연도: {missing}")

    return df
```

### 4. 데이터 불러오기

```python
# 파일 경로 설정
folder_path = r"C:\Users\SJ\Desktop\26-1\데이터마이닝\파주시아파트"
input_file = os.path.join(folder_path, "new_city.csv")

# CSV 파일 불러오기
df = pd.read_csv(input_file, encoding='utf-8-sig')
print(f"불러온 데이터: {len(df)}건")
```

### 5. CPI 변수 추가 및 결과 확인

```python
# CPI 열 추가
df = add_CPI(df)

# CPI 기술통계 확인
print("CPI 통계:")
print(df['CPI'].describe())
```

```python
# 연도별 CPI 평균 확인
print(df.groupby('계약연도')['CPI'].mean())
```

### 6. 결과 저장

```python
# CPI가 추가된 데이터를 동일 파일에 저장
output_file = os.path.join(folder_path, "new_city.csv")
df.to_csv(output_file, index=False, encoding='utf-8-sig')
print(f"저장 완료: {output_file}")
```
