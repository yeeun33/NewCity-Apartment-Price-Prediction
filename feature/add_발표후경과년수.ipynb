# **발표후경과년수 변수 생성**

### 1. 라이브러리 임포트

```python
import pandas as pd
import os
```

### 2. 신도시별 발표년도 정의

```python
# 신도시별 지구지정 고시 기준 발표년도
신도시_발표년도 = {
    '운정': 2003,
    '송도': 2003,
    '청라': 2003,
    '판교': 2003,
    '광교': 2005,
    '검단': 2006,
}
```

### 3. 발표후경과년수 파생변수 생성 함수 정의

```python
def add_발표후경과년수(df):
    # 계약년월 앞 4자리에서 계약연도 추출
    df['계약연도'] = df['계약년월'].apply(lambda x: int(str(x)[:4]))

    # 도시명을 기준으로 신도시 발표년도 조회하여 열 추가
    df['발표후경과년수'] = df['도시명'].apply(lambda x: 신도시_발표년도.get(x))

    # 딕셔너리에 등록되지 않은 도시가 있으면 안내 출력
    missing_mask = df['발표후경과년수'].isnull()
    if missing_mask.sum() > 0:
        missing = pd.unique(df.loc[missing_mask, '도시명'])
        print(f"발표년도 미등록 도시: {missing} — 딕셔너리에 추가해주세요.")

    # 발표후경과년수 = 계약연도 - 발표년도
    df['발표후경과년수'] = df['계약연도'] - df['발표후경과년수']

    return df
```

### 4. 데이터 불러오기

```python
# 파일 경로 설정
folder_path = r"C:\Users\SJ\Desktop\26-1\데이터마이닝\파주시아파트"
input_file = os.path.join(folder_path, "woonjeong_merged.csv")

# CSV 파일 불러오기 (한국어로 볼 수 있게)
df = pd.read_csv(input_file, encoding='utf-8-sig')
print(f"불러온 데이터: {len(df)}건")
```

### 5. 파생변수 생성 및 결과 확인

```python
# 발표후경과년수 열 추가
df = add_발표후경과년수(df)

# 기술통계 확인
print("발표후경과년수 통계:")
print(df['발표후경과년수'].describe())
```

```python
# 샘플 데이터 확인
print(df[['도시명', '계약년월', '계약연도', '발표후경과년수']].head(10))
```

### 6. 결과 저장

```python
# 파생변수가 추가된 데이터를 동일 파일에 저장
output_file = os.path.join(folder_path, "woonjeong_merged.csv")
df.to_csv(output_file, index=False, encoding='utf-8-sig')
print(f"저장 완료: {output_file}")
```
